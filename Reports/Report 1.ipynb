Entonces, si hay un aumento del gasto del gobierno, impacta directamente en el aumento de la demanda agregada (al haber más capacidad económica), lo que aumenta lo que implica un aumento en el ingreso-gasto en el nivel de producción. aunque no de forma inmediata pero es debido a que se tiende al equilibrio.  

Ahora, la justificación matemática:
Si ∆G_0 > 0  →  ¿∆Y?
o, considerando el multiplicador keynesiano,


$$ k > 0 $$

$$ Y = k (C_0 + I_0 + G_0 + X_0 - hr) $$

Pero, si no ha habido cambios en Co,Io Xo,h ni r , entonces:

 $$ ∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0 $$

$$∆Y = k (∆G_0)$$

Sabiendo que ∆G_0 > 0 y que k > 0, la multiplicación de un número positivo con un positivo dará otro positivo:

$$∆Y = (+)(+)$$

$$∆Y > 0$$
